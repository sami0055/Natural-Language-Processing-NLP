```python
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
from nltk.corpus import stopwords
import nltk

# Download NLTK stopwords
nltk.download('stopwords')

# Define stop words
stop_words = set(stopwords.words('english'))

story = []
for filename in os.listdir('data'):
    # Open the file with the correct encoding
    with open(os.path.join('data', filename), 'r', encoding='latin-1') as f:
        corpus = f.read()

    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        # Tokenize the sentence and remove stopwords
        filtered_sent = [word for word in simple_preprocess(sent) if word not in stop_words]
        story.append(filtered_sent)

# Display the first few sentences to verify the result
print(story[:5])
```

```python
story
```

```python
model=gensim.models.word2vec.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```
